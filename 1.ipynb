### Q3: Explain the use of the following operators using an example:

(i) **/** (Division)
```python
# Example of /
result = 10 / 3
print(result)  # Output: 3.3333333333333335
```
The `/` operator performs floating-point division, returning a float.

(ii) **%** (Modulus)
```python
# Example of %
remainder = 10 % 3
print(remainder)  # Output: 1
```
The `%` operator returns the remainder of the division.

(iii) **//** (Floor Division)
```python
# Example of //
floor_division = 10 // 3
print(floor_division)  # Output: 3
```
The `//` operator performs floor division, returning the largest integer less than or equal to the division result.

(iv) **\*\*** (Exponentiation)
```python
# Example of **
power = 2 ** 3
print(power)  # Output: 8
```
The `**` operator raises the left operand to the power of the right operand.

### Q5: Using a while loop, verify if number A is purely divisible by number B and if so then how many times it can be divisible.

```python
A = 100
B = 5
count = 0

while A % B == 0:
    A = A // B
    count += 1

print(f"Number can be divided {count} times by {B}")
```

### Q6: Create a list containing 25 int type data. Using for loop and if-else condition, print if the element is divisible by 3 or not.

```python
# Creating the list
numbers = list(range(1, 26))

# Using for loop and if-else condition
for number in numbers:
    if number % 3 == 0:
        print(f"{number} is divisible by 3")
    else:
        print(f"{number} is not divisible by 3")
```

### Q7: Explain mutable and immutable data types with examples.

- **Mutable Data Types:** These are data types whose value can be changed after they are created. Examples include lists, dictionaries, and sets.
    ```python
    # Example of mutable data type
    my_list = [1, 2, 3]
    my_list.append(4)
    print(my_list)  # Output: [1, 2, 3, 4]
    ```

- **Immutable Data Types:** These are data types whose value cannot be changed after they are created. Examples include strings, tuples, and integers.
    ```python
    # Example of immutable data type
    my_tuple = (1, 2, 3)
    try:
        my_tuple[0] = 4
    except TypeError as e:
        print(e)  # Output: 'tuple' object does not support item assignment
    ```
