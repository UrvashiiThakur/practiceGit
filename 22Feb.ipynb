### Q1: Create a Pandas Series that contains the following data: 4, 8, 15, 16, 23, and 42. Then, print the series.

```python
import pandas as pd

# Create the Pandas Series
data = [4, 8, 15, 16, 23, 42]
series = pd.Series(data)

# Print the series
print(series)
```

Output:
```
0     4
1     8
2    15
3    16
4    23
5    42
dtype: int64
```

### Q3: Create a Pandas DataFrame that contains the following data: Then, print the DataFrame.

```python
# Create a dictionary with the data
data = {
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12],
    'D': [13, 14, 15, 16]
}

# Convert the dictionary to a Pandas DataFrame
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

### Q4: What is ‘DataFrame’ in pandas and how is it different from pandas.Series? Explain with an example.

A **DataFrame** in pandas is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). It is similar to a spreadsheet or SQL table.

A **Series** in pandas is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.).

**Example:**

```python
# Series
series = pd.Series([1, 2, 3, 4, 5])
print("Series:")
print(series)

# DataFrame
data = {
    'Column1': [1, 2, 3, 4, 5],
    'Column2': ['A', 'B', 'C', 'D', 'E']
}
df = pd.DataFrame(data)
print("\nDataFrame:")
print(df)
```

Output:
```
Series:
0    1
1    2
2    3
3    4
4    5
dtype: int64

DataFrame:
   Column1 Column2
0        1       A
1        2       B
2        3       C
3        4       D
4        5       E
```

### Q5: What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

Some common functions to manipulate data in a Pandas DataFrame include:

- `df.head()`: Displays the first few rows of the DataFrame.
- `df.tail()`: Displays the last few rows of the DataFrame.
- `df.describe()`: Generates descriptive statistics.
- `df.drop(columns=['column_name'])`: Drops specified columns.
- `df.loc[]`: Access a group of rows and columns by labels or a boolean array.
- `df.iloc[]`: Access a group of rows and columns by integer position.
- `df.sort_values(by='column_name')`: Sorts the DataFrame by a specific column.

**Example:**

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'Score': [85, 62, 91, 70]
}
df = pd.DataFrame(data)

# Sort the DataFrame by 'Score'
sorted_df = df.sort_values(by='Score')
print(sorted_df)
```

### Q7: Create a DataFrame using multiple Series. Explain with an example.

```python
# Create individual Series
series1 = pd.Series([1, 2, 3], name='Column1')
series2 = pd.Series(['A', 'B', 'C'], name='Column2')

# Create a DataFrame from the Series
df = pd.concat([series1, series2], axis=1)

# Print the DataFrame
print(df)
```

Output:
```
   Column1 Column2
0        1       A
1        2       B
2        3       C
```

In this example, we created two Series and combined them into a DataFrame using the `pd.concat` function, which concatenates objects along a particular axis.
